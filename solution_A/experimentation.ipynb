```python
import pandas as pd

df = pd.read_csv('data/training_data/training_data/NLI/train.csv')
df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer

# Concatenate premise and hypothesis
df['text'] = df['premise'] + ' ' + df['hypothesis']

# Define pipeline with TfidfVectorizer and LogisticRegression
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', LogisticRegression(max_iter=1000))
])

# Define hyperparameters to search over
parameters = {
    'tfidf__max_features': [1000, 2000, 3000],
    'clf__C': [0.1, 1, 10]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(pipeline, parameters, cv=5, n_jobs=-1)
grid_search.fit(df['text'], df['label'])

# Print best parameters
print("Best Parameters:", grid_search.best_params_)
```

```python
eval_df = pd.read_csv('data/training_data/training_data/NLI/dev.csv')
eval_df
```

```python
eval_df.fillna('', inplace=True)
eval_pred = grid_search.predict(eval_df['premise'] + ' ' + eval_df['hypothesis'])
```

```python
from classes import evaluation

evaluation.evaluate(eval_df['label'], eval_pred)
```
